```python
import pandas as pd

df = pd.read_csv("../data/raw/iris.csv")
df.head()
df.info()
df.describe()
df['class'].value_counts()
```

```python
import matplotlib.pyplot as plt

df.hist(figsize=(10, 6), bins=15)

# save the figure in report/
plt.savefig("../reports/figures/iris_histogram.png", dpi=300, bbox_inches='tight')

plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='class', y='sepal length')
plt.title('Boxplot of Sepal Length by Iris Class')
plt.savefig("../reports/figures/iris_sepal_length_boxplot.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
sns.pairplot(df, hue='class')
plt.savefig("../reports/figures/iris_pairplot.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(10, 6))

sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
# plt.savefig("../reports/figures/iris_heatmap.png", dpi=300, bbox_inches='tight')
plt.show()
```
